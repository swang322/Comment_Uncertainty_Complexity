```python
# 使用最长公共前缀判断共祖
def common_parent_node(tree_numbers: list) -> int: 
    ans = ''	# 定义空字符串，准备用来存放最长公共子串
    for i in zip(*tree_numbers):
        if len(set(i)) == 1:	
            ans += i[0]
        else:
            break
    return len(ans)

df_mesh["TREENUBER Common"] = df_mesh["TREENUBER"].apply(lambda x:common_parent_node(eval(x)))
df_mesh
```
