```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scienceplots
import matplotlib.patches as mpatches
import matplotlib.ticker as ticker

plt.style.use(['science', 'nature'])
```

```python
# Data Preparation
df = pd.read_excel("./Data/Fig1_a.xlsx", sheet_name="Fig1.b_PubMed_timeline")
df.drop(0, inplace=True)
x = df["Year"]
y = df["Comment in_Count"]
y_rate = df["Com_percentage"]
```

```python
# Draw figure
fig, ax = plt.subplots()
fig.set_size_inches(6,4)
fig.set_dpi(300)


plt.figure(dpi=300)
ax.bar(x=x, height=y, color="#6495ED", label="Count")
ax.set_ylabel("Commented Studies(\#)", fontsize=12)
ax.set_xlabel("Year", fontsize=12)
ax.tick_params(axis="both", labelsize=12)

ax1 = ax.twinx()
sns.scatterplot(x=x, y=[u*100 for u in y_rate], markers='o', color='#B22222', edgecolor="none", ax=ax1)
sns.lineplot(x=x, y=[u*100 for u in y_rate], color='#B22222', ax=ax1, label=r"Rate(\%)")
ax1.set_ylabel(r"Commented Rate(\%)", fontsize=12)
ax1.set_ylim(0,6)
ax1.tick_params(axis="y", labelsize=12)

# 获取柱状图和折线图的图例对象
bars, labels1 = ax.get_legend_handles_labels()
lines, labels2 = ax1.get_legend_handles_labels()

# 合并图例对象
ax1.legend(bars + lines, labels1 + labels2, loc='upper left', fontsize=12)

plt.show()
```
