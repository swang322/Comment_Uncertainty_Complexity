```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scienceplots
import matplotlib.patches as mpatches
import matplotlib.ticker as ticker

plt.style.use(['science', 'nature'])
```

```python
df = pd.read_excel("./Data/Fig1_b.xlsx", sheet_name="PubMed_incidence")
```

```python
plt.figure(dpi=300, figsize=(4,3))
plt.plot(df["time_diff"], df["Incidence_rate_interjournal"], marker="o", label="Intra-journal")

plt.plot(df["time_diff"], df["Incidence_rate_acrosssjournal"], marker="o", label="Inter-journal")

plt.plot(df["time_diff"], df["Incidence_rate_all"], marker="o", label="All")


plt.xlabel("PubMed: Time Difference (in Month)", fontsize=10)
plt.ylabel("Incidence Rate", fontsize=10)
plt.xticks(list(range(1,13)))
ax = plt.gca()
ax.xaxis.set_minor_locator(ticker.MultipleLocator(base=1))
plt.ylim(0,1)
plt.xticks(size=10)
plt.yticks(size=10)
plt.legend(fontsize=10, bbox_to_anchor=(0.45, 0.5))
plt.show()
```
